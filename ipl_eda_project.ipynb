```python
matches_data['winner'].value_counts().plot(kind='bar',figsize=(10,6),color='lightblue')
plt.title("Matches won by the teams")
plt.xlabel("teams")
plt.ylabel('number of wins')
plt.xticks(rotation=45)
plt.grid(axis='y',linestyle='--',alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
matches_data['toss_decision'].value_counts().plot(kind='bar',color='coral')
plt.title('Toss_decisions')
plt.xlabel('decision')
plt.ylabel('number of times')
plt.xticks(rotation=0)
plt.grid(axis='y',linestyle=':',alpha=1)
plt.show()
```

```python
matches_data['city'].value_counts().plot(kind='barh',color='cyan')
plt.title("Matches per city")
plt.ylabel('city')
plt.xlabel('number of matches')
plt.grid(axis='y',linestyle='--',alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
matches_data['player_of_match'].value_counts().head(10).plot(kind='bar',color='plum')
plt.title('player of the match awards')
plt.xlabel("names")
plt.ylabel("awards number")
plt.xticks(rotation=45)
plt.show()
```

```python
matches_data['toss_decision'].value_counts().plot(kind='pie',autopct='%1.1f%%',startangle=90,colors=['mediumseagreen','red'])
plt.title('toss_decisions')
plt.ylabel('')
plt.show()
```
